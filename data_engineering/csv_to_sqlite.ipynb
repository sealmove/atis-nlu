The goal is to adjust the data from the csv files and insert them in the following relation database tables:
```sql
create table airports (code text primary key, name text, city text)
create table flights (from_airport_id text, dest_airport_id text, departure_time text, price real)
```

Clease data from `flight-data.csv`:

````python
import os
import pandas as pd

flights_dataset = 'flight-data.csv'
flights = pd.read_csv(flights_dataset)
flights = flights[['from_airport_code', 'dest_airport_code', 'departure_time', 'price']]
flights
````

Clease data from `airports-extended.dat` and insert them into the database:

````python
airports_dataset = 'airports-extended.dat'
airports = pd.read_csv(airports_dataset)
airports = airports.iloc[:, [4, 1, 2]]
airports.columns = ['code', 'name', 'city']
airports = airports[airports['code'] != '\\N']
airports
````

Create an sqlite database and insert the dataframes into it:

````python
from pathlib import Path
import sqlite3

Path('flights.db').touch()
conn = sqlite3.connect('flights.db')
c = conn.cursor()

flights.to_sql('flights', conn, index=False)
airports.to_sql('airports', conn, index=False)
````
